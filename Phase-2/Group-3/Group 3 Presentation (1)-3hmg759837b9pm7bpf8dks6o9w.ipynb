# **1.INTRODUCTION TO HYPOTHESIS TESTING** 

This is a statistical methods used to draw conclusions on a population based on sample data.
It helps determine whether there is enough evidence to reject a hypothesis(null hypothesis)

# Key Concepts in Hypothesis Testing

## 1.Hypothesis:
A statement about a population parameter (like mean or proportion) that we want to test.

Null Hypothesis (H₀): No effect, no difference, or the status quo.

Alternative Hypothesis (H₁ or Hₐ): There is an effect, a difference, or the new claim being tested.

## 2.P-Value:

The probability of getting the observed results, or something more extreme, if the null hypothesis is true.

Small p-value (< 0.05): Reject the null hypothesis (strong evidence for H₁).

Large p-value (> 0.05): Fail to reject the null hypothesis (not enough evidence).





## 3.Significance Level (α):

A threshold (usually 0.05) set before the test that determines how much risk you’re willing to take for rejecting a true null hypothesis.

## 4.T-Statistic or Z-Statistic:

A value calculated from the sample data that measures how far the observed result is from the null hypothesis.


# **2.INTRODUCTION TO EXPERIMENTAL DESIGN** 

This is the process of planning experiments to test hypothesis effectively 

It ensures that all subsequent results are reliable, conclusions are valid and that sources of error or bias are minimized

## 1.steps of scientific method

**1.Make an observation**- This is made on the phenomena you wish to test

**2.Examine the research** - Examine the already existing research that may help in answering the question

**3.Form a hypothesis** - Develop a testable explanation. Alternative hypothesis is what you intend to test and Null hypothesis is the opposite which is assumed to be true

**4.Conduct an experiment**- Design and conduct an experiment to test your hypothesis. A well-structured experiment is one that accounts for all of the mistakes and randomness that could give you false signals relating to the effect of an intervention.

**5.Analyse experimental results**- During this step, you will tease out relationships, filter out noise, and try to determine if something that happened is statistically significant or not.

**6.Draw conclusions**- use the results of the analysis step to evaluate whether you believe the hypothesis was correct or not! 

## 2. Foundations of a sound experiment
### 1. Presence of control group

This is a group that receives no treatment or intervention and acts as a baseline to compare with the treatment group.

Without a control group, it’s impossible to determine if observed results are due to the intervention or would have happened naturally.

Control groups must be sampled from the same population as the treatment group to ensure valid comparisons.
This is achieved through Random Controlled Trials- Participants are randomly assigned to either the control group or the treatment group.

To further reduce bias **single blind** or **double blind** studies are conducted.

**Single blind** - Where only the participants are not aware of which group they are at

**Double blind** - Both administrators and participants aren't aware of their groups

### 2.Appropriate Sample Techniques

Since Data Scientists rarely get to work with whole populations, sampling is of essence. This are the 4 assumptions

1.Sample should be independent- individual data items should have no connection or correlation so as not to skew results

2.Sample is collected randomly- Each data item has an equal chance of being selected

3.Sample is approximately normally distributed- 

The normal distribution assumption is that the sampling distribution of the mean is normal. That is, if you took a sample, calculated its mean, and then you took another (independent) sample (from the same population) and got its mean (and repeated this an infinite number of times), then the distribution of the values that you wrote down would always be a perfect bell curve

4.Appropriate Sample size- The sample size should be large enough to detect meaningful differences or effects.
Larger sample sizes reduce sampling error and improve statistical power.

### 3.Reproducibility
This one is a big one, and it represents a bit of a crisis in some parts of the scientific community right now. Good scientific experiments have Reproducible Results! 

This means that if someone else follows the steps you outline for your experiment and performs it themselves, they should get pretty much the same results as you did (allowing for natural variance and randomness) 

## **3.Conducting a T-test**

### What is a T-test

A T-test is a method used to compare averages (means) between groups to see if the difference is real or just by chance.

For example:

Do two classrooms have the same average scores?

Did people improve after training (before vs. after)?

Is the average weight of apples different from 100 grams? 

## When to run a T-test
1. The standard deviation is unknown
2. When the sample size is small

## Types of T-Tests

**One-Sample T-Test** : Compare a group's mean to a single number.
Example: Is the average exam score equal to 70?

**Two-Sample T-Test** : Compare the means of two independent groups.
Example: Do boys and girls have the same average height?

**Paired T-Test** : Compare the means of the same group before and after a change.
Example: Did students score better after training compared to before? 

## Assumptions for a T-Test

Before running a T-test, check these conditions:

**Normality**: Data should follow a roughly normal distribution (bell-shaped curve).

**Independence**: Observations (data points) must not depend on each other.

**Equal Variance**: For two-sample T-tests, the variance (spread) in the two groups should be similar.

**Numeric and continuous**: 

## Steps for performing a T-test

1. Set up null and alternative hypothesis

2. Choose a significance level 

3. Calculate the T-Statistic

4. Determine the critical T value(Find the rejection region)

5. Compare the T-value with critical T-value to determine whether we can reject the null hypothesis

## Sample Question:

Acme Ltd. wants to improve sales performance. Past sales data indicate that the average sale was 100 dollars per transaction. After training the sales force, recent sales data (from a random sample of 25 salesmen) is shown below.

      [122.09, 100.64, 125.77, 120.32, 118.25,  
        96.47, 111.4 ,  80.66, 110.77, 111.14, 
        102.9, 114.54,  88.09,  98.59,  87.07, 
       110.43, 101.9 , 123.89,  97.03, 116.23, 
        108.3, 112.82, 119.57, 131.38, 128.39]

```python
## Import the packages
import numpy as np
from scipy import stats 
import math

# For visualizing distributions - optional 
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data listed above
sample = np.array([122.09, 100.64, 125.77, 120.32, 118.25,  96.47, 111.4 , 80.66,
       110.77, 111.14, 102.9 , 114.54,  88.09,  98.59,  87.07, 110.43,
       101.9 , 123.89,  97.03, 116.23, 108.3 , 112.82, 119.57, 131.38,
       119])
```

```python
# Population mean (μ)
# (known from past data)
mu = 100
mu

```

```python
# Sample mean (x̄) using NumPy array method mean()
x_bar = sample.mean()
x_bar
```

```python
# Sample standard deviation (sigma) using NumPy function std()
sigma = np.std(sample, ddof=1)
sigma
```

```python
# Sample size (n)
n = len(sample)
# Degrees of Freedom
df = n-1
df

```

```python
# Difference between sample mean and population mean
diff = x_bar - mu
diff
```

The sample contains 25 observations, having a mean of 109.5456
and a standard deviation (sigma) of 13.339, with 24 degrees of freedom.

The difference between sample and population means is 9.55.

```python
## Checking if the sample is roughly normally distributed

sns.set(color_codes=True)
sns.histplot(sample, kde=True, bins=5, color='darkblue');
```

## Step 1: Establish Null and Alternative Hypotheses



H₀ : The null hypothesis is that the training did not increase sales,

H₀ : μ ≤ 100


Our alternative hypothesis should address the expected change in the sales performance

i.e. the sales performance has increased and the mean of sales post-training is greater than 100.


Hₐ : μ > 100

## Step 2: Choose a Significance Level (Alpha)

The significance level, also denoted as alpha or α, is the probability of rejecting the null hypothesis when it is true.

For example, a significance level of 0.05 indicates a 5% risk of concluding that a difference exists when there is no actual difference

We will assume you can assume a α of 0.05

## Step 3: Calculate the t-statistic


```python
t = (x_bar -  mu)/(sigma/np.sqrt(n))
t
```

```python
# Set up figure and axes
fig, ax = plt.subplots(figsize=(8,5))

# Generate data for plotting the PDF
# (1) generate points on the x axis between -5 and 5:
x = np.linspace(-5, 5, 200)
# (2) use stats.t.pdf to get values on the probability density
# function for the t-distribution
y = stats.t.pdf(x, df, 0, 1)

# Plot the PDF as a line graph
ax.plot(x, y, color='darkblue', label="t-distribution PDF")

# Plot a vertical line for our measured difference in sales t-statistic
ax.axvline(t, color='red', linestyle='--', lw=5,label='observed t-statistic')

ax.legend();
```

## Step 4: Calculate Critical Value (Find Rejection Region)


```python
t_crit = stats.t.ppf(1 - 0.05, df=24)
t_crit
```

```python
# Set up figure and axes
fig, ax = plt.subplots(figsize=(8,5))

# Plot the PDF as a line graph
# (x and y were created in previous plotting code)
ax.plot(x, y, color='darkblue', label="t-distribution PDF")

# Plot a vertical line for our critical t-value
ax.axvline(t_crit,color='green',linestyle='--',lw=4,label='critical t-value')

# Fill area under line graph where x value is creater than critical t
ax.fill_betweenx(y,x,t_crit,where=x > t_crit,color="gray",label="rejection region")

ax.legend();
```

The critical t-value marks the boundary of the rejection region. 

Because this is a one-tailed test, we can simply say that the rejection region is anything greater than the critical t-statistic.

If our sample t-statistic falls into the rejection region, we can reject the null hypothesis!

## Step 5: Compare Sample t-value with Critical t-value: Can We Reject the Null Hypothesis?


```python
# Set up figure and axes
fig, ax = plt.subplots(figsize=(8,5))

# Plot the PDF as a line graph
# (x and y were created in previous plotting code)
ax.plot(x, y, color='darkblue', label="t-distribution PDF")

# Plot a vertical line for our critical t-value
ax.axvline(t_crit,color='green',linestyle='--',lw=4,label='critical t-value')

# Fill area under line graph where x value is creater than critical t
ax.fill_betweenx(y,x,t_crit,where=x > t_crit,color="gray",label="rejection region")

# Plot a vertical line for our measured difference in sales t-statistic
ax.axvline(t, color='red', linestyle='--', lw=5,label='observed t-statistic')

ax.legend();
```

## Observations

Our observed t-statistic is greater than the critical t-value. 

which means that it is in the rejection region,

which means that we can reject the null hypothesis at a significance level of 0.05,

which means that there was a statistically significant increase in sales performance for those who received the training (at our specified significance level)
